```python
import pandas as pd
import numpy as np
```

```python
movies= pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
movies= movies.merge(credits,on= "title")
movies.dropna(inplace= True)
```

```python
movies.isnull().sum()
```

```python
movies= movies[["movie_id","title","cast","crew","genres","keywords","overview"]]
```

```python
movies.head()
```

```python
import ast 
def fetch_genres(row):
    genres= []
    for genre in ast.literal_eval(row):
        genres.append(genre["name"])
    return genres
```

```python
movies["genres"]=movies["genres"].apply(fetch_genres)
```

```python
movies.head()
```

```python
import ast 
def fetch_cast(row):
    cast= []
    counter= 0 
    for x in ast.literal_eval(row):
        if counter < 3:
            cast.append(x["name"])
            counter += 1
        else:
            break
    return cast
```

```python
movies["cast"]=movies["cast"].apply(fetch_cast)
```

```python
movies.head()
```

```python
import ast
def convert(obj):
    crew=[]
    for i in ast.literal_eval(obj):
        if i["job"]== "Director":
          crew.append(i["name"])
    return crew
```

```python
movies["crew"]=movies["crew"].apply(convert)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
import ast 
def fetch_keywords(row):
    keywords= []
    counter= 0 
    for x in ast.literal_eval(row):
        if counter < 3:
            keywords.append(x["name"])
            counter += 1
        else:
            break
    return keywords
```

```python
movies["keywords"]=movies["keywords"].apply(fetch_keywords)
```

```python
movies.head()
```

```python
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" ","") for i in x ])
movies["crew"] = movies["crew"].apply(lambda x : [i.replace(" ","") for i in x ])
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" ","") for i in x ])
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" ","") for i in x ])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] +  movies["cast"] +  movies["crew"] +  movies["genres"] +  movies["keywords"] 
```

```python
movies.head()
```

```python
df=movies[["movie_id", "title", "tags"]]
```

```python
df.head()
```

```python
df["tags"] = df["tags"].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
df["tags"] = df["tags"].apply(lambda x :x.lower())
```

```python
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemm(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df["tags"]= df["tags"].apply(stemm)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=  CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors= cv.fit_transform(df["tags"]).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
def recommend(movie):
    index=df[df["title"] == movie].index[0]
    distances= sorted(list(enumerate(similarity[index])) ,reverse= True, key = lambda x:x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend("Inception")
```
